```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
telcom.columns[1:]
```

```python
for k in telcom.columns[1:]:
    print(telcom[k].dtypes)
```

```python
for k in telcom.columns[1:]:
    if telcom[k].dtype == object:
        le.fit(telcom[k])
        telcom[k] = le.transform(telcom[k]).astype('int64')
```

```python
telcom
```

```python
for k in telcom.columns[1:]:
    print(telcom[k].dtypes)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = telcom.drop(['Churn'], axis = 1)
```

```python
y = telcom['Churn']
```

```python
x
```

```python
y
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size= 0.2, random_state =0)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
telrandtree = RandomForestClassifier(n_estimators = 100, random_state = 0, max_depth = 3, class_weight = 'balanced')
telrandtree.fit(x_train.iloc[ : , 1:], y_train)
y_pred = telrandtree.predict(x_test.iloc[:,1:])
```

```python
from sklearn import tree
```

```python
from IPython.display import Image
```

```python
from io import StringIO
import pydotplus
```

```python
dot_data = StringIO()
tree.export_graphviz(telrandtree.estimators_[99], out_file=dot_data, filled = True, rounded = True, special_characters= True, feature_names = ['tenure', 'PhoneService', 'MultipleLines', 'InternetService', 'OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies', 'Contract', 'PaperlessBilling', 'PaymentMethod', 'MonthlyCharges', 'TotalCharges', 'gender', 'SeniorCitizen', 'Partner', 'Dependents'], class_names =['Yes', 'No'])
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())
graph.write_png('Random Forest')
Image(graph.create_png())
```
